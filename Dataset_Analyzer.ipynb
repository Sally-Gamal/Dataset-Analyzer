```python
import pandas as pd
import numpy as np

def dataset_analyzer():
    print("DatasetAnalyzer is running...\n")

    # Dataset file name (must be in the same directory)
    file_name = "global_student_migration.csv"

    try:
        df = pd.read_csv(file_name)

        # ---------------- Data Overview ----------------
        print("üìä Data Overview")
        print("-" * 40)

        print(f" Rows: {df.shape[0]}")
        print(f" Columns: {df.shape[1]}")
        print("-" * 40)

        # Display first 5 rows
        print("\nüîù First 5 Rows:")
        print(df.head())
        print("-" * 40)

        # Display column information and data types
        print("\n Columns Info:")
        df.info()
        print("-" * 40)

        # ---------------- Missing Values ----------------
        print("\n‚ùó Missing Values:")
        missing = df.isnull().sum()
        print(missing[missing > 0] if missing.sum() > 0 else "0Ô∏è‚É£ No missing values")
        print("-" * 40)
        # ---------------- Duplicates ----------------
        duplicated_count = df.duplicated().sum()
        print("\n Duplicated Rows:")
        print(f"{duplicated_count} duplicated rows" if duplicated_count > 0 else "0Ô∏è‚É£ No duplicated rows")
        print("-" * 40)
        # ---------------- Data Cleaning ----------------
        print("\nüßπ Cleaning data...")

        # Remove duplicated rows only
        df = df.drop_duplicates()

        # ---------------- Save Clean Data ----------------
        output_file = "global_student_migration_cleaned.csv"
        df.to_csv(output_file, index=False)

        print("\n‚úÖ Data analysis & cleaning completed")
        print(f" Clean file saved as: {output_file}")

    except Exception as e:
        # Handle file reading errors
        print("‚ùå Error while reading the file")
        print(e)


if __name__ == "__main__":
    dataset_analyzer()
```
